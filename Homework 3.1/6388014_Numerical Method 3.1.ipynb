```python
#q2
import math
import matplotlib.pyplot as plt
import numpy as np
accuracy = 0.5
def f(x):
    return (x**3) - (6 * (x**2)) + (4*x) + 12
array=[-2,6]
array1=[2.5,3.5]  
array2=[4.5,5.5]
def inn(array,accuracy):
    x = np.arange(array[0],array[1],accuracy)
    n_intervals = len(x) - 1
    for k in range(0,n_intervals):
        xl = x[k]
        xu = x[k+1]
        if ((f(xl) * f(xu) < 0)):
            array =[]
            array.append(xl)
            array.append(xu)
            xr = (xl+xu)/2
            break;
    return xl,xu;
def bisection(xl,xu):
    epa = 100
    eps = 0.005
    prev_xr = 0
    while (epa > eps):
        print('searching [%3.4f,%3.4f] ... ' %(xl,xu), end = ' ') 
        xr = (xl + xu)/2
        if (f(xl) * f(xr) < 0):
            xu = xr
        else:
            xl = xr
        epa = abs( (xr - prev_xr)/xr) *100
        prev_xr = xr
        print('updated to [%3.4f,%3.4f] xr = %3.4f epa = %3.4f' %(xl,xu,xr,epa))
    return xr
xl,xu=inn(array,accuracy)
xl,xu=inn(array1,accuracy)
xl,xu=inn(array2,accuracy)
print("A range between [-1.5000,-1.0000]:")
myroot1=bisection(*inn(array,accuracy))
print('The estimated root is %3.4f' %(myroot1))
print("A range between [2.5000,3.0000]:")
myroot2=bisection(*inn(array1,accuracy))
print('The estimated root is %3.4f' %(myroot2))
print("A range between [4.5000,5.0000]:")
myroot3=bisection(*inn(array2,accuracy))
print('The estimated root is %3.4f' %(myroot3))
```
